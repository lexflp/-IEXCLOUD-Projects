```python
import pandas as pd
import requests
import os
```

```python
base_url = 'https://cloud.iexapis.com/stable'
sandbox_url = 'https://sandbox.iexapis.com/stable'

token = os.environ.get('CLOUD_TOKEN')
params = {'token': token}
```

```python
def time_series (_dates):
    endpoint = f'{base_url}/time-series/REPORTED_FINANCIALS/{_dates}/10-Q?from=2011-01-01&to=2020-06-01'
    #if _range:
        #endpoint += f'/{_range}'
    #elif _date:
        #endpoint += f'/date/{_date}'
    
    resp = requests.get(endpoint, params=params)
    resp.raise_for_status()
    return pd.DataFrame(resp.json())
```

```python
AAPL= time_series('AAPL')
```

```python
writer = pd.ExcelWriter('AAPLtimeseries.xlsx', engine='xlsxwriter')

AAPL.to_excel(writer, sheet_name='AAPL Timeseries')

header_template = writer.book.add_format(
        {
            'font_color': '#ffffff',
            'bg_color': '#135485',
            'border': 1
        }
    )

string_template = writer.book.add_format(
        {
            'bg_color': '#DADADA',
            'border': 1
        }
    )

dollar_template = writer.book.add_format(
        {
            'num_format':'$0.00',
            'bg_color': '#DADADA',
            'border': 1
        }
    )

percent_template = writer.book.add_format(
        {
            'num_format':'0.0%',
            'bg_color': '#DADADA',
            'border': 1
        }
    )
```

```python
#Format the header of the spreadsheet
writer.sheets['AAPL Timeseries'].conditional_format('A1:QJ1', 
                             {
                                'type':     'cell',
                                'criteria': '<>',
                                'value':    '"None"',
                                'format':   header_template
                                }
                            )

#Format the 'Ticker' and 'Company Name' columns
writer.sheets['AAPL Timeseries'].conditional_format('A2:B11', 
                             {
                                'type':     'cell',
                                'criteria': '<>',
                                'value':    '"None"',
                                'format':   string_template
                                }
                            )

#Format the 'Stock Price' column
writer.sheets['AAPL Timeseries'].conditional_format('C2:C11', 
                             {
                                'type':     'cell',
                                'criteria': '<>',
                                'value':    '"None"',
                                'format':   dollar_template
                                }
                            )

#Format the 'Dividend Yield' column
writer.sheets['AAPL Timeseries'].conditional_format('D2:D11', 
                             {
                                'type':     'cell',
                                'criteria': '<>',
                                'value':    '"None"',
                                'format':   percent_template
                                }
                            )
```

```python
writer.save()
```
