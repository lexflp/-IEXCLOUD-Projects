```python
import pandas as pd
import requests
import os
```

```python
base_url = 'https://cloud.iexapis.com/stable'
sandbox_url = 'https://sandbox.iexapis.com/stable'

token = os.environ.get('CLOUD_TOKEN')
params = {'token': token}
```

```python
def top_10 (_list):
    endpoint = f'{base_url}/stock/market/list/{_list}'
    #if _range:
        #endpoint += f'/{_range}'
    #elif _date:
        #endpoint += f'/date/{_date}'
    
    resp = requests.get(endpoint, params=params)
    resp.raise_for_status()
    return pd.DataFrame(resp.json())
```

```python
top_10= top_10('mostactive')
```

```python
writer = pd.ExcelWriter('top10mostactive.xlsx', engine='xlsxwriter')

top_10.to_excel(writer, sheet_name='Most Active Stocks')

header_template = writer.book.add_format(
        {
            'font_color': '#ffffff',
            'bg_color': '#135485',
            'border': 1
        }
    )

string_template = writer.book.add_format(
        {
            'bg_color': '#DADADA',
            'border': 1
        }
    )

dollar_template = writer.book.add_format(
        {
            'num_format':'$0.00',
            'bg_color': '#DADADA',
            'border': 1
        }
    )

percent_template = writer.book.add_format(
        {
            'num_format':'0.0%',
            'bg_color': '#DADADA',
            'border': 1
        }
    )
```

```python
#Format the header of the spreadsheet
writer.sheets['Most Active Stocks'].conditional_format('A1:BD1', 
                             {
                                'type':     'cell',
                                'criteria': '<>',
                                'value':    '"None"',
                                'format':   header_template
                                }
                            )

#Format the 'Ticker' and 'Company Name' columns
writer.sheets['Most Active Stocks'].conditional_format('A2:B11', 
                             {
                                'type':     'cell',
                                'criteria': '<>',
                                'value':    '"None"',
                                'format':   string_template
                                }
                            )

#Format the 'Stock Price' column
writer.sheets['Most Active Stocks'].conditional_format('C2:C11', 
                             {
                                'type':     'cell',
                                'criteria': '<>',
                                'value':    '"None"',
                                'format':   dollar_template
                                }
                            )

#Format the 'Dividend Yield' column
writer.sheets['Most Active Stocks'].conditional_format('D2:D11', 
                             {
                                'type':     'cell',
                                'criteria': '<>',
                                'value':    '"None"',
                                'format':   percent_template
                                }
                            )
```

```python
writer.save()
```
